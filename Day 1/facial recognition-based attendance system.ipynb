```python
import cv2
from deepface import DeepFace
import pandas as pd
from datetime import datetime
import os
import time

# ===========================================
# 📁 FACE DATABASE
# ===========================================
database = {
    "Emmanual": r"D:\Ipcs\facial recognition-based attendance system\MANU.jpg",
    "Kevin": r"D:\Ipcs\facial recognition-based attendance system\person2.jpg"
}

# ===========================================
# 📄 ATTENDANCE FILE SETUP
# ===========================================
attendance_file = "attendance.csv"

if not os.path.exists(attendance_file):
    df = pd.DataFrame(columns=["Name", "Time"])
    df.to_csv(attendance_file, index=False)

# ===========================================
# 🕒 FUNCTION TO MARK ATTENDANCE
# ===========================================
def mark_attendance(name):
    df = pd.read_csv(attendance_file)
    if name not in df["Name"].values:
        now = datetime.now()
        time_string = now.strftime("%H:%M:%S")
        new_entry = pd.DataFrame([[name, time_string]], columns=["Name", "Time"])
        df = pd.concat([df, new_entry], ignore_index=True)
        df.to_csv(attendance_file, index=False)
        print(f"[INFO] ✅ Attendance marked for {name}")
    else:
        print(f"[INFO] ⏳ {name} already marked present.")

# ===========================================
# 🎥 START WEBCAM
# ===========================================
cap = cv2.VideoCapture(0)
time.sleep(2)  # Give webcam time to warm up

if not cap.isOpened():
    print("[ERROR] ❌ Webcam not accessible.")
    exit()
else:
    print("[INFO] 🎥 Webcam started. Press 'q' to quit.")

frame_count = 0

# ===========================================
# 🔁 MAIN LOOP
# ===========================================
while True:
    ret, frame = cap.read()
    if not ret:
        print("[ERROR] ❌ Failed to grab frame.")
        break

    frame_count += 1

    # Process every 10th frame to speed up
    if frame_count % 10 != 0:
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        continue

    # Save current frame temporarily
    cv2.imwrite("live_face.jpg", frame)

    recognized = False

    # Compare against database
    for name, db_img in database.items():
        if not os.path.exists(db_img):
            print(f"[WARNING] ⚠️ Image for {name} not found: {db_img}")
            continue

        try:
            result = DeepFace.verify(
                img1_path="live_face.jpg",
                img2_path=db_img,
                enforce_detection=False,
                model_name="VGG-Face"
            )

            if result["verified"]:
                mark_attendance(name)
                recognized = True
                cv2.putText(frame, f"{name} ✅", (50, 50),
                            cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 3)
                break

        except Exception as e:
            print(f"[ERROR] {e}")

    if not recognized:
        cv2.putText(frame, "❌ Unrecognized Face!", (50, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 0, 255), 3)

    # Show the frame
    cv2.imshow("Face Attendance - Press 'q' to Quit", frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# ===========================================
# 🧹 CLEANUP
# ===========================================
cap.release()
cv2.destroyAllWindows()
print("[INFO] 👋 Program closed.")
```
